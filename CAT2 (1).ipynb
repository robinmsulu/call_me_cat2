```python
# Load the student enrollment CSV file
enrollment_df = pd.read_csv('student_enrollment_data_v2.csv')
```

```python
# Display the first few rows of the dataset
print("First few rows of the dataset:")
print(enrollment_df.head())
```

```python
# Summary statistics for GPA and attendance rate
print("\nSummary statistics for GPA and Attendance Rate:")
print(enrollment_df[['gpa', 'attendance_rate']].describe())
```

```python
# Visualize GPA distribution
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
plt.figure(figsize=(12, 6))
sns.histplot(data=enrollment_df, x='gpa', bins=20, kde=True, color='blue')
plt.title('GPA Distribution')
plt.xlabel('GPA')
plt.ylabel('Frequency')
plt.show()
```

```python
# Visualize attendance rate distribution
plt.figure(figsize=(12, 6))
sns.histplot(data=enrollment_df, x='attendance_rate', bins=10, kde=False, color='green')
plt.title('Attendance Rate Distribution')
plt.xlabel('Attendance Rate (%)')
plt.ylabel('Frequency')
plt.show()
```

```python
# Analyze GPA by socioeconomic status
plt.figure(figsize=(12, 6))
sns.boxplot(data=enrollment_df, x='socioeconomic_status', y='gpa', palette='coolwarm')
plt.title('GPA by Socioeconomic Status')
plt.xlabel('Socioeconomic Status')
plt.ylabel('GPA')
plt.show()
```

```python
# Highlight students with GPA < 2.5 or attendance rate < 70%
low_performers = enrollment_df[
    (enrollment_df['gpa'] < 2.5) | (enrollment_df['attendance_rate'] < 70)
]

print("\nStudents with low performance (GPA < 2.5 or Attendance Rate < 70%):")
print(low_performers[['student_id', 'gpa', 'attendance_rate']])
```
